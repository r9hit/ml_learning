```python
# single feature (Pclass) linear_regression

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Loading cleaned dataset
df = pd.read_csv("titanic_dataset/cleaned_train.csv")

# Select a single feature(Pclass)
X = df["Pclass"].values

# target
y = df["Fare"].values
```

```python
# Compute means
x_mean = X.mean()
y_mean = y.mean()

# Calculate slope (w)
numerator = np.sum((X - x_mean) * (y - y_mean))
denominator = np.sum((X - x_mean) ** 2)
w = numerator / denominator

# Calculate intercept (b)
b = y_mean - w * x_mean

print(f"Slope (w): {w:.4f}")
print(f"Intercept (b): {b:.4f}")
```

```python
# Predict on training data
y_pred = w * X + b
```

```python

# Plot
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', label='Predicted Line')
plt.xlabel("Pclass")
plt.ylabel("Fare")
plt.title("Single Feature Linear Regression : training data")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Evaluation metrics (manual)
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def mae(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

print(f"Training data Manual MSE: {mse(y, y_pred):.2f}")
print(f"Training data Manual MAE: {mae(y, y_pred):.2f}")
```

```python
# Prediction on testing data

#loading cleaned test data
df_test = pd.read_csv("titanic_dataset/cleaned_test.csv")

#single feature from test data
X_test = df_test["Pclass"].values

#target in test data
y_test = df_test["Fare"].values
```

```python
#prediction on testing data
y_test_pred = w * X_test + b

# Save predictions
df_test["Fare_Predicted"] = y_test_pred
```
